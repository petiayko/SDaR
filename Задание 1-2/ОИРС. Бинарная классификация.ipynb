```python
import numpy as np
import pandas as pd
```

# Загрузка данных

```python
DF_NAME = 'Химический анализ ДСП'
DF_FILENAME_XLSX = DF_NAME + '.xlsx'
DF_FILENAME_CSV = DF_NAME + '.csv'
```

```python
df = pd.read_csv(DF_FILENAME_CSV)
df
```

```python
df.info()
```

# Удаление нечисловых параметров

```python
df.drop(['Дата продукта', 'МС Код', 'МС Вариант', 'МС Наименование', 'МС Условный аналог', 'Дата выпуска',
         'НТД выплавки', '№ ДСП', 'Дата отбора пробы', '№  пробы'], axis=1, inplace=True)
```

# Группировка данных по номеу плавки с усреднением

```python
df = df.groupby('Номер плавки').mean()
df = df.reset_index()
df = df.fillna(0)
df.drop(['Номер плавки'], axis=1, inplace=True)
```

```python
for i in range(len(df)):
    df.at[i, 'Качество'] = 1 if df.at[i, 'Качество'] >= 0.5 else 0
df
```

```python
corr = df.drop('Качество', axis=1).corr()
corr.style.background_gradient(cmap='coolwarm')
```

# Подготовка к моделированию

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

```python
def train_and_predict(model_t, d_train, d_test, t_train, t_test):
    model_t.fit(d_train, t_train)
    prediction = model_t.predict(d_test)
    return f1_score(t_test, prediction), accuracy_score(t_test, prediction)
```

```python
df_data = df.drop('Качество', axis=1)
target = df['Качество']

data_train, data_test, target_train, target_test = train_test_split(df_data, target, test_size=0.3, random_state=42)
```

```python
df_answer = pd.DataFrame({
    'Score': ['F1', 'Accuracy'],
    'KNN': [None, None],
    'SVC': [None, None],
    'RF': [None, None],
    'MLP': [None, None],
    'ANS': [None, None],
})
df_answer = df_answer.set_index('Score')
```

# Моделирование разными алгоритмами

```python
algorithms = [('knn', KNeighborsClassifier(n_neighbors=15)),
              ('svc', svm.SVC()),
              ('rf', RandomForestClassifier()),
              ('mlp', MLPClassifier())]

for name, clssifier in algorithms:
    df_answer[name.upper()] = train_and_predict(clssifier, data_train, data_test, target_train, target_test)

df_answer['ANS'] = train_and_predict(StackingClassifier(estimators=algorithms[:3],
                                                        final_estimator=algorithms[3][1]), data_train, data_test, target_train, target_test)
```

# Построение перцептрона

```python
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Dense
```

```python
K.clear_session()
model = Sequential()

model.add(Dense(15, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(25, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy',
              metrics=[keras.metrics.F1Score(threshold=0.5)])
```

```python
model.fit(data_train, target_train, epochs=100, batch_size=64,
          validation_data=(data_test, target_test))
model.summary()
```

```python
history = pd.DataFrame(model.history.history)
for i in range(len(history)):
    history.at[i, 'f1_score'] = float(history.at[i, 'f1_score'][0])
    history.at[i, 'val_f1_score'] = float(history.at[i, 'val_f1_score'][0])
history = history.astype({'f1_score': 'float64', 'val_f1_score': 'float64'})
```

```python
pd.options.plotting.backend = 'plotly'
```

```python
fig = history.plot(title='Потери', template='simple_white',
                   labels=dict(index='Эпоха', value='Значение',
                               variable='option'))
fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)
fig.show()
```
